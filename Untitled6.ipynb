```python
plt.scatter(ship['PassengerId'], ship['Age'],color='purple')
plt.xlabel('PassengerId')
plt.ylabel('Age')
plt.title('PassengerId vs Age')
plt.show()
```

```python
survival_counts = ship['Survived'].value_counts()
```

```python
labels = ['Not Survived', 'Survived']
values = survival_counts.values
plt.bar(labels, values,color='purple')
plt.xlabel('Survival')
plt.ylabel('Count')
plt.title('Survival Count in Titanic Dataset')
plt.show()
```

```python
sex_age_counts = ship.groupby(['Sex', 'Age']).size().reset_index(name='Count')
labels = sex_age_counts['Sex'].astype(str) + ', ' + sex_age_counts['Age'].astype(int).astype(str)
values = sex_age_counts['Count']
plt.bar(labels, values,color='purple')
plt.xlabel('Sex, Age')
plt.ylabel('Count')
plt.title('Count of Passengers by Sex and Age')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
sns.set(style='darkgrid')
plt.subplots(figsize = (15,8))
ax=sns.countplot(x='Sex',data=ship,hue='Survived',edgecolor=(0,0,0),linewidth=2)
ship.shape
## Fixing title, xlabel and ylabel
plt.title('Passenger distribution of survived vs not-survived',fontsize=25)
plt.xlabel('Gender',fontsize=15)
plt.ylabel("# of Passenger Survived", fontsize = 15)
labels = ['Female', 'Male']
#Fixing xticks.
plt.xticks(sorted(ship.Survived.unique()),labels)
## Fixing legends
leg = ax.get_legend()
leg.set_title('Survived')
legs=leg.texts
legs[0].set_text('No')
legs[1].set_text('Yes')
```

```python
plt.subplots(figsize=(10,8))
sns.kdeplot(ship.loc[(ship['Survived'] == 0),'Pclass'],shade=True,color='r',label='Not Survived')
ax=sns.kdeplot(ship.loc[(ship['Survived'] == 1),'Pclass'],shade=True,color='b',label='Survived' )

labels = ['First', 'Second', 'Third']
plt.xticks(sorted(ship.Pclass.unique()),labels)
```

```python
plt.subplots(figsize=(15,10))
ax=sns.kdeplot(ship.loc[(ship['Survived'] == 0),'Fare'],color='r',shade=True,label='Not Survived')
ax=sns.kdeplot(ship.loc[(ship['Survived'] == 1),'Fare'],color='b',shade=True,label='Survived' )
plt.title('Fare Distribution Survived vs Non Survived',fontsize=25)
plt.ylabel('Frequency of Passenger Survived',fontsize=20)
plt.xlabel('Fare',fontsize=20)
```
